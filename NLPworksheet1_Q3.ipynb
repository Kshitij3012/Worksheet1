```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['Kshitij', 'Krishna', 'Sahaj', 'Vedant', 'Vipul', 'Taksh', 'Jay', 'Aditya', 'Shivam', 'Chinmay'],
    'Subject': ['English', 'Maths', 'Science', 'English', 'Maths', 'Science', 'English', 'Maths', 'Science', 'English'],
    'Score': np.random.randint(50, 101, size=10),
    'Grade': ''  })

print(df)
```

```python
g = []

for score in df['Score']:
    if score >= 90:
        g.append('A')
    elif score >= 80:
        g.append('B')
    elif score >= 70:
        g.append('C')
    elif score >= 60:
        g.append('D')
    else:
        g.append('F')

df['Grade'] = g
print(df)
```

```python
print(df.sort_values(by='Score', ascending=False))
```

```python
print(df.groupby('Subject')['Score'].mean())
```

```python
def pandas_filter_pass(dataframe):
    return dataframe[(dataframe['Grade'] == 'A') | (dataframe['Grade'] == 'B')]

print(pandas_filter_pass(df))
```
