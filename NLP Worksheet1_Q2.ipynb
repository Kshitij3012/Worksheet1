```python
print(np.max(a))
print(np.min(a))
print(np.mean(a))
```

```python
a[a < 5] = a[a < 5]**2
print(a)
```

```python
import numpy as np

def numpy_alternate_sort(arr):
    d = arr.size
    arr = np.sort(arr)
    i=0

    for i in range(d):
     if i%2 == 0:
      arr[i:d] = np.sort(arr[i:d])

     else:
      arr[i:d] = np.sort(arr[i:d])[::-1]

    return arr

print(numpy_alternate_sort(a))
```
